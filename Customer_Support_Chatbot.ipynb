```python
import requests
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer
```

```python
model_name = "EleutherAI/gpt-neo-2.7B"
model = AutoModelForCausalLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
qa_pipeline = pipeline('text-generation', model=model, tokenizer=tokenizer)
```

```python
# Sample knowledge base specific to Amazon
knowledge_base = {
    "order tracking": "You can track your order status in your Amazon account under 'Your Orders'. Enter your order number to see the latest updates.",
    "returns": "Amazon allows returns within 30 days of delivery for most items. Go to 'Your Orders' and select 'Return or Replace Items' to start a return.",
    "product availability": "You can check the availability of a product on its detail page. If it's out of stock, you can choose to be notified when it's back in stock.",
    "payment methods": "Amazon accepts all major credit cards, debit cards, and Amazon Pay. You can manage your payment methods in 'Your Account' under 'Payment Options'.",
    "shipping options": "Amazon offers various shipping options including standard, expedited, and priority shipping. You can choose your preferred option at checkout."
}
```

```python
# Function to generate response using LLM
def generate_response(prompt):
    response = qa_pipeline(prompt, max_length=100, do_sample=True, temperature=0.7)
    return response[0]['generated_text']
```

```python
# Chatbot function
def chatbot(query):
    # Check knowledge base first
    for key, value in knowledge_base.items():
        if key in query.lower():
            return value

   # If no good match, use LLM to generate a response
    prompt = f"Customer query: {query}\nChatbot response:"
    llm_response = generate_response(prompt)
    return llm_response
```

```python
queries = [
    "How do I track my order on Amazon?",
    "What is the return policy on Amazon?",
    "How can I check product availability?",
    "What payment methods does Amazon accept?",
    "What are the shipping options available?"
]
```

```python
for query in queries:
    response = chatbot(query)
    print(f"Query: {query}\nResponse: {response}\n")
```
